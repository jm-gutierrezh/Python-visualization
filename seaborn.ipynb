Statistical data visualization in Python.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
sns.lmplot(x='total_bill', y='tip', hue='time', col='sex', row='smoker', palette='Set1', data=tips)
```
