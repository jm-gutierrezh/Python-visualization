# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Basic plots
x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)

#Line plot
plt.plot(x, y, 'teal')
plt.show()
```

## Plotting multiple graphs

* Common-axis graphs
* Different axes
* Subplots

```python
#Common-axis
x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)
z = np.cos(x)
w = np.sin(2*x)

plt.plot(x, y, 'red')
plt.plot(x, z, 'blue')
plt.show()
```

```python
#Different axes - plt.axes([x_lower_left_position, y_lower_left_position, width, height])
plt.axes([0, 0, 1, 1.1])
plt.plot(x, y, 'red')
plt.axes([1.1, 0, 1, 0.5])
plt.plot(x, z, 'blue')
plt.axes([1.1, 0.6, 1, 0.5])
plt.plot(x, w, 'green')
plt.show()
```

```python
#Subplots - plt.subplot(nrows, ncols, nsubplot)
#nsubplot ordered row-wise from top left and indexed from 1

f = plt.figure(figsize=(15,3))
ax1 = f.add_subplot(1, 3, 1)
ax2 = f.add_subplot(1, 3, 2)
ax3 = f.add_subplot(1, 3, 3)
ax1.plot(x, y, 'red')
ax2.plot(x, z, 'blue')
ax3.plot(x, w, 'green')
plt.show()
```

## Controlling axis options

* Setting axis limits

```python
#Setting axis limits
f = plt.figure(figsize=(15,3))
ax1 = f.add_subplot(1, 3, 1)
ax2 = f.add_subplot(1, 3, 2)
ax1.plot(x, y, 'red')
ax2.plot(x, y, 'red')
ax2.set_xlim(0, 3.2)
ax2.set_ylim(0, 1.1)
plt.show()
```

## Legends, annotations and styles

* Markers: https://matplotlib.org/api/markers_api.html
* Colors: https://www.quackit.com/css/css_color_codes.cfm

```python
plt.plot(x, y, color='red', label='sin(x)')
plt.plot(x, z, color='blue', label='cos(x)')
plt.legend(loc='lower left')
plt.annotate(s='sin', xy=(1.8, 1), xytext=(2.4, 0.8), arrowprops={'color':'green'})
plt.annotate(s='cos', xy=(1.2, 0.25), xytext=(1.8, 0), arrowprops={'color':'green'})
plt.show()
```
